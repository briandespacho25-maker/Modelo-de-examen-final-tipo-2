```python
import pandas as pd
import numpy as np
import random

# --- 1. Definición de datos base para la generación aleatoria ---

# Apellidos y Nombres comunes (fragmentos para combinar)
apellidos = [
    'García', 'Rodríguez', 'González', 'Fernández', 'López', 'Martínez',
    'Díaz', 'Pérez', 'Sánchez', 'Romero', 'Gómez', 'Juárez', 'Flores',
    'Acuña', 'Benítez', 'Cabrera', 'Navarro', 'Torres', 'Vargas', 'Ramos'
]
nombres = [
    'Juan', 'María', 'Carlos', 'Laura', 'Pedro', 'Ana', 'Diego', 'Sofía',
    'Lucas', 'Valentina', 'Mateo', 'Camila', 'Facundo', 'Lucía', 'Santiago',
    'Emilia', 'Nicolás', 'Martina', 'Franco', 'Agustina'
]

# Materias escolares secundarias argentinas
materias = [
    'Matemática', 'Lengua y Literatura', 'Historia', 'Geografía',
    'Biología', 'Física', 'Química', 'Inglés', 'Educación Ciudadana',
    'Educación Física', 'Arte Visuales', 'Tecnología de la Información'
]

N_FILAS = 100

# --- 2. Generación de las columnas de datos ---

# Apellido y Nombre
# Combina un nombre y un apellido aleatorio para cada fila
apellido_nombre = [
    f"{random.choice(apellido)} {random.choice(nombre)}"
    for apellido, nombre in zip(np.random.choice(apellidos, N_FILAS), np.random.choice(nombres, N_FILAS))
]

# nro_dni
# Genera DNI simulados de 8 dígitos (entre 10.000.000 y 55.000.000)
# Usa np.random.randint para generar un array de números enteros
nro_dni = np.random.randint(10000000, 55000000, N_FILAS)

# materia escolar
# Elige una materia aleatoria con reemplazo
materia_escolar = np.random.choice(materias, N_FILAS)

# Notas Trimestrales (1er, 2do, 3er)
# Genera notas entre 1 y 10. Usa np.random.uniform para notas con decimales o np.random.randint para enteras.
# Elegiremos enteros para simular notas cerradas comunes.
# np.random.randint(low=1, high=11) -> va de 1 (incluido) a 10 (incluido)
nota_1er_trimestre = np.random.randint(1, 11, N_FILAS)
nota_2do_trimestre = np.random.randint(1, 11, N_FILAS)
nota_3er_trimestre = np.random.randint(1, 11, N_FILAS)

# Faltas a clases
# Genera un número de faltas entre 0 y 20
faltas_a_clases = np.random.randint(0, 21, N_FILAS)

# --- 3. Creación del DataFrame de pandas ---

datos = {
    'Apellido y Nombre': apellido_nombre,
    'nro_dni': nro_dni,
    'materia escolar': materia_escolar,
    'nota 1er Trimestre': nota_1er_trimestre,
    'nota 2do Trimestre': nota_2do_trimestre,
    'nota 3er Trimestre': nota_3er_trimestre,
    'Faltas a clases': faltas_a_clases
}

df_estudiantes = pd.DataFrame(datos)

# --- 4. Visualización del resultado (Opcional: Muestra las primeras 5 filas) ---

# print(df_estudiantes.head())
# print(f"\nEl DataFrame tiene {len(df_estudiantes)} filas.")
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

# --- CÓDIGO DE GENERACIÓN DE DATOS (Mismo del paso anterior) ---

apellidos = ['García', 'Rodríguez', 'González', 'Fernández', 'López', 'Martínez', 'Díaz', 'Pérez', 'Sánchez', 'Romero', 'Gómez', 'Juárez', 'Flores', 'Acuña', 'Benítez', 'Cabrera', 'Navarro', 'Torres', 'Vargas', 'Ramos']
nombres = ['Juan', 'María', 'Carlos', 'Laura', 'Pedro', 'Ana', 'Diego', 'Sofía', 'Lucas', 'Valentina', 'Mateo', 'Camila', 'Facundo', 'Lucía', 'Santiago', 'Emilia', 'Nicolás', 'Martina', 'Franco', 'Agustina']
materias = ['Matemática', 'Lengua y Literatura', 'Historia', 'Geografía', 'Biología', 'Física', 'Química', 'Inglés', 'Educación Ciudadana', 'Educación Física', 'Arte Visuales', 'Tecnología de la Información']
N_FILAS = 100

apellido_nombre = [f"{random.choice(apellido)} {random.choice(nombre)}" for apellido, nombre in zip(np.random.choice(apellidos, N_FILAS), np.random.choice(nombres, N_FILAS))]
nro_dni = np.random.randint(10000000, 55000000, N_FILAS)
materia_escolar = np.random.choice(materias, N_FILAS)
nota_1er_trimestre = np.random.randint(1, 11, N_FILAS)
nota_2do_trimestre = np.random.randint(1, 11, N_FILAS)
nota_3er_trimestre = np.random.randint(1, 11, N_FILAS)
faltas_a_clases = np.random.randint(0, 21, N_FILAS)

datos = {
    'Apellido y Nombre': apellido_nombre,
    'nro_dni': nro_dni,
    'materia escolar': materia_escolar,
    'nota 1er Trimestre': nota_1er_trimestre,
    'nota 2do Trimestre': nota_2do_trimestre,
    'nota 3er Trimestre': nota_3er_trimestre,
    'Faltas a clases': faltas_a_clases
}
df_estudiantes = pd.DataFrame(datos)

# --- INICIO DEL ANÁLISIS Y GENERACIÓN DEL GRÁFICO ---

# 1. Calcular el promedio anual por registro (estudiante-materia)
df_estudiantes['Promedio Anual'] = df_estudiantes[['nota 1er Trimestre', 'nota 2do Trimestre', 'nota 3er Trimestre']].mean(axis=1)

# 2. Agrupar por materia y calcular los agregados
df_agrupado = df_estudiantes.groupby('materia escolar').agg(
    Promedio_General=('Promedio Anual', 'mean'), # Promedio de notas
    Total_Faltas=('Faltas a clases', 'sum')       # Suma total de faltas
).reset_index()

# 3. Ordenar por Promedio General de mayor a menor
df_agrupado = df_agrupado.sort_values(by='Promedio_General', ascending=False)

# 4. Crear el Gráfico Combinado con doble eje Y
fig, ax1 = plt.subplots(figsize=(12, 6))

# Eje Principal (ax1) - Promedio de Notas (Barras)
color_notas = 'tab:blue'
ax1.set_xlabel('Materia Escolar')
ax1.set_ylabel('Promedio General de Notas (1-10)', color=color_notas)
barras = ax1.bar(df_agrupado['materia escolar'], df_agrupado['Promedio_General'], color=color_notas, alpha=0.6, label='Promedio de Notas')
ax1.tick_params(axis='y', labelcolor=color_notas)
ax1.set_ylim(0, 10) # Fija el límite del eje de notas

# Eje Secundario (ax2) - Total de Faltas (Línea)
ax2 = ax1.twinx() # Crea un segundo eje Y que comparte el mismo eje X
color_faltas = 'tab:red'
ax2.set_ylabel('Total de Faltas Acumuladas', color=color_faltas)
linea = ax2.plot(df_agrupado['materia escolar'], df_agrupado['Total_Faltas'], color=color_faltas, marker='o', label='Total de Faltas')
ax2.tick_params(axis='y', labelcolor=color_faltas)

# Ajustes de visualización y leyendas
plt.title('Promedio General de Notas vs. Total de Faltas por Materia')
fig.tight_layout() # Para asegurar que no se solapen las etiquetas
ax1.set_xticklabels(df_agrupado['materia escolar'], rotation=45, ha="right") # Rotar etiquetas para mejor lectura

# Unir leyendas de ambos ejes
handles = [barras] + linea
labels = ['Promedio de Notas', 'Total de Faltas']
ax1.legend(handles, labels, loc='upper right')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
